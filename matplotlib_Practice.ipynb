Multiple Line Graphs = Pyplot style

```python
x = np.linspace(0,2,100)
y1 = 0.5*x
y2 = 0.5*x**2
y3 = 0.5*x**3
```

```python
plt.plot(x,y1,label='Linear')
plt.plot(x,y2,label='Quadratic')
plt.plot(x,y3,label='Cubic')
plt.legend()
plt.ylabel('Y')
plt.xlabel('X')
plt.title('Graphs')
```

Multiple Line Graphs - OOP Style

```python
hf, ha = plt.subplots()
hp1, = ha.plot(x,y1)
hp2, = ha.plot(x,y2)
hp3, = ha.plot(x,y3)
ha.set_ylabel('Y')
ha.set_xlabel('X')
ha.set_title('Graphs')
hp1.set_label('Linear')
hp2.set_label('Quadratic')
hp3.set_label('Cubic')
ha.legend()
```

```python
hp1.set_color('k')
hp2.set_color('b')
hp3.set_color('r')
ha.legend()
ha.grid()
hf
```
